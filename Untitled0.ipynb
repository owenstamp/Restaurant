```python
import pandas as pd

# Load Excel file
df = pd.read_excel('sales_data.csv.xlsx')

# Show the first 5 rows
df.head()
```

```python
# prompt: “Write Python code (for Colab) to show summary statistics and data types for a CSV called sales_data.csv.”

# Show summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Show data types of each column
print("\nData Types:")
print(df.info())
```

```python
# prompt: “Write code to create a correlation heatmap using seaborn for numerical columns.”

import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns
df_numeric = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

```python
# prompt: “Write code to check for missing or null values in the dataset.”

import matplotlib.pyplot as plt
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Visualize missing values (optional, but helpful)
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Value Heatmap')
plt.show()
```
